# **Project Description**

The main goal of this project is to develop an efficient memory-based directory system using the STM32F413ZHT6 microcontroller in direct interaction with the W25Q32JV memory. This memory-based directory system will enable organized and effective management of data stored in the flash memory.

The project design involves generating a configuration based on the Serial Peripheral Interface (SPI) protocol to facilitate communication between the microcontroller and an external flash memory.

Key focus areas include:

- Microcontroller Configuration: Establishing parameters for effective communication between the STM32F413ZHT6 and the W25Q32JV memory.
- Memory Interaction: Implementing procedures to read, write, and manage the flash memory from the microcontroller.
- Directory System Development: Designing an organized directory structure to facilitate access and management of stored data.







# **Theoretical Framework**




## W25Q32JV
The Serial Flash Memory W25Q32JV is designed to provide a compact storage solution suitable for systems with limitations in terms of space, pins, and power. As part of the 25Q series, this memory device surpasses the capabilities of typical Serial Flash memories, offering flexibility and performance. It is suitable for various applications, including code shadowing to RAM, direct code execution from Dual/Quad SPI (XIP), and storing voice, text, and data.

Operating within a power supply range of 2.7V to 3.6V, the W25Q32JV boasts low power consumption, with standby current consumption as minimal as 1μA. The array is organized into 16,384 programmable pages, each comprising 256 bytes. Programming can occur in increments of up to 256 bytes at a time. Page erasure can be performed at different granularities: 4KB sector erase, 32KB block erase, 64KB block erase, or full chip erase. The device features 1,024 erasable sectors and 64 erasable blocks.

A notable feature is its support for the Standard Serial Peripheral Interface (SPI) and Dual/Quad I/O SPI. The W25Q32JV allows clock frequencies of up to 133MHz, equivalent to 266MHz for Dual I/O and 532MHz for Quad I/O when using the Fast Read Dual/Quad I/O instructions. These transfer rates surpass those of standard 8-bit and 16-bit parallel Flash memories. Continuous Read Mode allows efficient memory access with only 8 instruction overhead cycles needed to read a 24-bit address, supporting true eXecute In Place (XIP) operation.

The device includes additional control features such as a hold pin, write protection pin, and programmable write protection with upper or lower block control. It complies with JEDEC standard manufacturer and device IDs and includes a 64-bit Unique Serial Number alongside three 256-byte Security Registers. Collectively, these features provide extensive control flexibility and security options for various applications.

## STM32F413ZHT6
The STM32F413ZHT6 microcontroller stands as a key component in the realm of embedded systems, offering a set of specialized features for efficient connection with external flash memories like the W25Q. These features are crucial to ensure smooth and effective communication between the microcontroller and the external memory, enabling efficient storage and retrieval of data in advanced embedded applications.

This microcontroller, anchored in STMicroelectronics' STM32 family, stands out for its powerful ARM Cortex-M4 core, providing an optimal balance between performance and energy efficiency. Designed to operate in a voltage range of 2.7V to 3.6V, the STM32F413ZHT6 effectively handles battery-powered systems or conventional power sources.

A key aspect of its functionality is the integration of an SPI interface, an essential requirement for communication with external flash memories. This Serial Peripheral Interface protocol facilitates data transfer between the microcontroller and the W25Q flash memory. The SPI communication speed and the microcontroller's ability to adapt to the specifications of the flash memory are crucial to ensuring efficient system operation. Managing the Chip Select (CS) signal is another fundamental element. Through a designated GPIO pin as Chip Select, the STM32F413ZHT6 can select and activate the external flash memory during read and write operations. This process is essential for precisely and timely directing flash memory operations.

Additionally, it is crucial to consider the microcontroller's ability to leverage advanced features of the flash memory, such as Fast Read modes and eXecute In Place (XIP) capability. These capabilities enable more efficient data reading and direct code execution from the flash memory, respectively.

## Guide to Connecting a Microcontroller to a Device Using the SPI Protocol.

*Requirements:*

Microcontroller with SPI compatibility.
SPI Device (Slave).
Wiring (typically jumper wires).
Power supply if required for the device.

*Connection Steps:*

1. Identify SPI Pins: On your microcontroller, locate the specific pins related to the SPI protocol, which usually include:
- SCK (Serial Clock): Clock line.
- MISO (Master In Slave Out): Slave-to-master data line.
- MOSI (Master Out Slave In): Master-to-slave data line.
- SS/CS (Slave Select/Chip Select): Line used to selecct the communication slave.

2. Pin Connection:
Use jumper cables to connect the microcontroller's pins to the corresponding pins on the slave device. Ensure that the connections are precise and match the data, clock, and slave selection lines.

3. Power Supply:
If the slave device requires an external power supply, make sure to provide it according to the device's specifications.

4. Microcontroller Configuration:
In your microcontroller code, configure essential SPI parameters such as:
- Clock speed
- Phase and polarity
- Adjust these parameters based on the libraries or development environment you are using. (In this case, we will use HAL.)

5.  Slave Selection:
Before initiating communication with a particular slave, enable the corresponding SS/CS line. This will signal to the slave device that it is ready for communication.

6. Data Transmission and Reception:
Use the functions or commands provided by your microcontroller to send and receive data through the SPI protocol. Typically, you will write data to the output register (MOSI) and read data from the input register (MISO) to achieve bidirectional communication.

7. Disable the Slave:
Once you have completed communication with a particular slave, deactivate the corresponding SS/CS line to indicate that communication with that device has finished.

8. Repeating the Process:
If you need to communicate with multiple slave devices, repeat steps 5-7, enabling and disabling the SS/CS lines as needed.

9. Completion and Cleanup:
After all communications are finished, make sure to properly close the SPI connections and perform any required cleanup in your code.

10. Testing and Troubleshooting:
Conduct thorough testing to confirm that the SPI communication is functioning as expected. If you encounter difficulties, use debugging tools and refer to the documentation for your devices and microcontroller.

## What is a PCB?

A Printed Circuit Board, commonly known as a PCB, is an essential component in modern electronics, found in a wide array of devices ranging from computers and monitors to mobile phones. Essentially, a PCB serves as a physical substrate for the installation and connection of electronic and electrical components. These components, including chips, capacitors, diodes, and resistors, are mounted on the PCB and interconnected through extremely fine copper conductor tracks. The PCB acts as a vital link that enables effective communication between these components.

The process of creating a PCB is complex and detailed. It begins with the design of the PCB using specialized software, where the necessary electrical tracks are traced, and the arrangement of components is established. Subsequently, screening and photolithography are performed, creating masks that indicate the component locations. The inner layers of the PCB are imprinted with conductor tracks through a photosensitive process, and the design integrity is verified through automated inspection.

The primary substrate of PCBs is typically a combination of fiberglass reinforced with ceramic materials, resins, and non-conductive plastics. PCBs also undergo hole drilling, hole metallization, and lamination processes to ensure connectivity between layers. The outer track layer is created through a similar process, followed by electroplating to deposit copper and tin on the conductor tracks.

Solder mask and legend are applied to facilitate component soldering and provide crucial information about the PCB. Finally, industrial-level component soldering is precisely carried out using robotic arms, followed by final tests to ensure the PCB's proper functionality. The invention and evolution of PCBs have been instrumental in the electronic devices revolution, providing an innovative method for connecting elements without the need for electrical wires and defining the foundation of modern electronics.


# **Varn Algorithm**

A research team form UCSD, Carolina Fernández, Ismael González, Paul H. Siegel, mentored by Simeng Zheng, Andrew Tan. Studied the following aspects.

The Varn algorithm addresses the problem of constructing minimum redundancy prefix codes for discrete noiseless channels. It focuses on situations where the costs of code letters are unequal, but the encoded symbols are assumed to be equally probable.

The Varn algorithm, also known as ACE (Arbitrary Symbol Cost and Equal Code Word Probability), was designed to solve two classes of problems:

1. Problem C1: Involves building an optimal prefix code that must also be exhaustive. A code is considered exhaustive if, for any pair of code symbols, a sequence of symbols is a prefix of a code word if and only if the other sequence is also a prefix.

2. Problem C2: Similar to Problem C1, but without the restriction that the optimal prefix code must be exhaustive.

The algorithm's approach involves using a graphical representation of prefix codes through trees. Each terminal node of a tree represents a code word, and each outgoing branch of a node is labeled with one of the symbols from the code alphabet. The optimal solution is constructed by expanding these trees, following certain rules to minimize the total code cost, where the cost is defined as the sum of the costs of all code words in the tree.

The Varn algorithm executes the following steps to solve the problem:

1. An optimal exhaustive prefix code is built for a larger number of words than necessary.
2. The most costly code words are eliminated until reaching the desired code size, paying attention to the number of distinct prefixes to ensure optimality.
3. The code with the lowest total cost is selected as the optimal solution.

This algorithm is significant because it provides an optimal solution for constructing prefix codes under constraints of arbitrary symbol costs, something not directly achievable with more traditional methods like the Huffman algorithm. Additionally, Varn's approach is applicable to a broader range of encoding problems, especially where symbol costs are not uniform.

# **Programming in the CubeIDE Environment.**

To determine if the microcontroller can recognize the memory, the slave device has its own factory identifier, constrained by a hexadecimal value. Connecting the identifier is achieved by sending a specific identification read command to the device. By reading a response containing the memory ID, it can be ensured that the microcontroller recognizes the memory's presence.

As previously outlined, reading and writing are crucial for addressing one of the main functionalities in this project. To achieve this, functions were defined to enable the aforementioned operations. The following functions are listed:

- W25Q_Reset(): This function resets the flash memory chip. It is responsible for sending the necessary command sequence to reset the memory.

- W25Q_Read/FastRead(): These operations perform reading in the memory chip. This function specifies the location from which the reading starts in the flash memory. 'startPage' refers to the page number from where the reading starts, 'offset' signifies the offset within the page where the reading starts, 'size' indicates the size of the data to be read, and finally, 'rData' holds a pointer to the buffer where the read data is stored.

- W25Q_Write_Page(): This function executes operations related to writing in the flash memory chip, thereby writing data to specific locations. The function takes four parameters: first, the page number where the writing starts; secondly, the 'offset', referring to the offset within the page to begin writing; thirdly, 'size', corresponding to the size of the data to be written (the number of characters); and finally, 'data', the data stored in the memory.

- W25Q_Erase_Sector: It receives a parameter indicating the sector number to be erased in the flash memory.


**Considerations with page offset**: It was mentioned that 'offset' determines the initial position within a memory page. In the hypothetical case where writing to the memory is required again, an offset calculation needs to be performed. After writing data to memory, the offset is updated by adding the length of the character string being sent. This ensures that the next writing occurs at a different position within the page, preventing the overwrite of previous data.

void W25Q_Reset(void)
{
    uint8_t tData[2];
    tData[0] = 0x66; // Con esta instrucción nos permite resetear televise
    tData[1] = 0x99; // Reset
    csLOW();
    SPI_Write(tData, 2);
    csHIGH();
    HAL_Delay(100);
}

uint32_t W25Q_ReadID(void)
{
    uint8_t tData = 0x9F;
    uint8_t rData[3];
    csLOW();
    SPI_Write(&tData, 1);
    SPI_Read(rData, 3);
    csHIGH();
    return ((rData[0] << 16) | (rData[1] << 8) | rData[2]);
}

void W25Q_Read (uint32_t startPage, uint8_t offset, uint32_t size, uint8_t *rData)
{
	uint8_t tData[5];
	uint32_t memAddr = (startPage*256) + offset;

	if (numBLOCK<512)   // Chip Size<256Mb
	{
		tData[0] = 0x03;  // enable Read
		tData[1] = (memAddr>>16)&0xFF;  // MSB of the memory Address
		tData[2] = (memAddr>>8)&0xFF;
		tData[3] = (memAddr)&0xFF; // LSB of the memory Address
	}
	else
	{
		tData[0] = 0x13;  // Read Data with 4-Byte Address
		tData[1] = (memAddr>>24)&0xFF;  // MSB of the memory Address
		tData[2] = (memAddr>>16)&0xFF;
		tData[3] = (memAddr>>8)&0xFF;
		tData[4] = (memAddr)&0xFF; // LSB of the memory Address
	}

	csLOW();  // pull the CS Low
	if (numBLOCK<512)
	{
		SPI_Write(tData, 4);  // send read instruction along with the 24 bit memory address
	}
	else
	{
		SPI_Write(tData, 5);  // send read instruction along with the 32 bit memory address
	}

	SPI_Read(rData, size);  // Read the data
	csHIGH();  // pull the CS High
}

void W25Q_FastRead (uint32_t startPage, uint8_t offset, uint32_t size, uint8_t *rData)
{
	uint8_t tData[6];
	uint32_t memAddr = (startPage*256) + offset;

	if (numBLOCK<512)   // Chip Size<256Mb
	{
		tData[0] = 0x0B;  // enable Fast Read
		tData[1] = (memAddr>>16)&0xFF;  // MSB of the memory Address
		tData[2] = (memAddr>>8)&0xFF;
		tData[3] = (memAddr)&0xFF; // LSB of the memory Address
		tData[4] = 0;  // Dummy clock
	}
	else
	{
		tData[0] = 0x0C;  // Fast Read with 4-Byte Address
		tData[1] = (memAddr>>24)&0xFF;  // MSB of the memory Address
		tData[2] = (memAddr>>16)&0xFF;
		tData[3] = (memAddr>>8)&0xFF;
		tData[4] = (memAddr)&0xFF; // LSB of the memory Address
		tData[5] = 0;  // Dummy clock
	}

	csLOW();  // pull the CS Low
	if (numBLOCK<512)
	{
		SPI_Write(tData, 5);  // send read instruction along with the 24 bit memory address
	}
	else
	{
		SPI_Write(tData, 6);  // send read instruction along with the 32 bit memory address
	}

	SPI_Read(rData, size);  // Read the data
	csHIGH();  // pull the CS High
}

void write_enable(void)
{
    uint8_t tData = 0x06;
    csLOW();
    SPI_Write(&tData, 1);
    csHIGH();
    W25Q_Delay(5);
}

void write_disable(void)
{
    uint8_t tData = 0x04;
    csLOW();
    SPI_Write(&tData, 1);
    csHIGH();
    W25Q_Delay(5);
}

uint32_t bytestowrite (uint32_t size, uint16_t offset)
{
	if ((size+offset)<256) return size;
	else return 256-offset;
}

void W25Q_Erase_Sector(uint16_t numsector)
{
    uint8_t tData[6];
    uint32_t memAddr = numsector * 16 * 256; // Each sector contains 16 pages * 256 bytes

	write_enable();
            uint32_t indx = 0;
	if (numBLOCK<512)   // Chip Size<256Mb
	{
		tData[0] = 0x02;  // page program
		tData[1] = (memAddr>>16)&0xFF;  // MSB of the memory Address
		tData[2] = (memAddr>>8)&0xFF;
		tData[3] = (memAddr)&0xFF; // LSB of the memory Address

		indx = 4;
	}
	else
	{
		tData[0] = 0x12;  // page program with 4-Byte Address
		tData[1] = (memAddr>>24)&0xFF;  // MSB of the memory Address
		tData[2] = (memAddr>>16)&0xFF;
		tData[3] = (memAddr>>8)&0xFF;
		tData[4] = (memAddr)&0xFF; // LSB of the memory Address

		indx = 5;
	}

    W25Q_Delay(450); // 450ms delay for sector erase

    write_disable();
}

void W25Q_Write_Page (uint32_t page, uint16_t offset, uint32_t size, uint8_t *data)
{
	uint8_t tData[266];
	uint32_t startPage = page;
	uint32_t endPage  = startPage + ((size+offset-1)/256);
	uint32_t numPages = endPage-startPage+1;

	uint16_t startSector  = startPage/16;
	uint16_t endSector  = endPage/16;
	uint16_t numSectors = endSector-startSector+1;
	for (uint16_t i=0; i<numSectors; i++)
	{
		//W25Q_Erase_Sector(startSector++);
	}

	uint32_t dataPosition = 0;

	// write the data
	for (uint32_t i=0; i<numPages; i++)
	{
		uint32_t memAddr = (startPage*256)+offset;
		uint16_t bytesremaining  = bytestowrite(size, offset);
        uint32_t indx = 0;

        write_enable();

        if (numBLOCK < 512) // Chip Size < 256Mb
        {
            tData[0] = 0x02; // page program
            tData[1] = (memAddr >> 16) & 0xFF; // MSB of the memory Address
            tData[2] = (memAddr >> 8) & 0xFF;
            tData[3] = (memAddr) & 0xFF; // LSB of the memory Address

            indx = 4;
        }
        else
        {
            tData[0] = 0x12; // page program with 4-Byte Address
            tData[1] = (memAddr >> 24) & 0xFF; // MSB of the memory Address
            tData[2] = (memAddr >> 16) & 0xFF;
            tData[3] = (memAddr >> 8) & 0xFF;
            tData[4] = (memAddr) & 0xFF; // LSB of the memory Address

            indx = 5;
        }

		uint16_t bytestosend  = bytesremaining + indx;

		for (uint16_t i=0; i<bytesremaining; i++)
		{
			tData[indx++] = data[i+dataPosition];
		}

		csLOW();
		SPI_Write(tData, bytestosend);
		csHIGH();

		startPage++;
		offset = 0;
		size = size-bytesremaining;
		dataPosition = dataPosition+bytesremaining;

		W25Q_Delay(5);
		write_disable();

	}

	// Esta función espera hasta que la escritura en la memoria flash haya finalizado
	void W25Q_WaitForWriteEnd(void)
	{
	    uint8_t cmd = 0x05; // Comando para leer el registro de estado
	    uint8_t status;

	    csLOW();
	    SPI_Write(&cmd, 1); // Enviar el comando para leer el registro de estado

	    // Leer el registro de estado y comprobar el bit de ocupado/buscando
	    do {
	        SPI_Read(&status, 1);
	    } while (status & 0x01); // 0x01 es el bit de ocupado/buscando en el registro de estado

	    csHIGH();
	}

}

```


# **Conclusions**

# Indentified challenges

We explained that the project did not progress and that tests were conducted.

Through the CubeIDE environment, a perspective shift was used while debugging the code to observe a sequence of variable updates, ensuring the writing and reading of the subject. Initially, the initial challenge was identifying the memory ID. Once the memory was recognized, the subsequent challenge was reading and writing. However, there was no real-time knowledge to ascertain whether the signals were effectively being implemented when executing writing commands. It was observed that signals were reaching the SPI device; however, it was determined that these signals (through the MISO, MOSI, SCK, CS channels) were not arriving correctly according to the timing diagrams in the slave device datasheet. Consequently, it is concluded that it was necessary to interact with the visualization of these signals, as it felt like working without explicitly seeing the behaviors generated by both devices. For future users intending to create their directory system, it is determined that they must have the appropriate equipment to understand these metrics and also consider evaluating the necessary configurations by the microcontroller.

# General conclusion

For our project, we have made significant progress working with the SPI protocol in connection with the W25Q32JV flash memory. However, there is a crucial component that is still pending implementation to further optimize our system: the integration of the Varn algorithm. Our intention with implementing the Varn algorithm in this specific context is to maximize the efficiency and lifespan of the flash memory. This algorithm enables us to create optimal shaping codes that minimize memory wear while maximizing storage capacity and data reliability transmitted through the SPI protocol. Implementing the Varn algorithm involves several steps, including document compression, calculating symbol probabilities based on symbol costs and compression ratios, and constructing a specific Varn code for our W25Q32JV flash memory. By implementing the Varn algorithm, we aim to achieve significant improvements in the flash memory's lifespan, minimizing wear during data write and read operations, which is crucial for optimal performance and enhanced reliability in our SPI protocol-based system connected to the W25Q32JV flash memory.

# **References**

González, A. G. (2022). ¿Cómo funciona el protocolo SPI? Panama Hitek. https://panamahitek.com/como-funciona-el-protocolo-spi/
Campbell, S. (s.f.). BASICS OF THE SPI COMMUNICATION PROTOCOL. https://www.circuitbasics.com/basics-of-the-spi-communication-protocol/

alldatasheet.com. (n.d.). W25Q32JV pdf, W25Q32JV Description, W25Q32JV Datasheet, W25Q32JV view ::: ALLDATASHEET  ::: https://pdf1.alldatasheet.com/datasheet-pdf/view/932078/WINBOND/W25Q32JV.html
STM32F413ZH - STMicroelectronics. (n.d.). STMicroelectronics. https://www.st.com/en/microcontrollers-microprocessors/stm32f413zh.html
Castillo, J. A. (2023, October 9). Qué es una PCB o Placa de Circuito Impreso. Uso, cómo se fabrica. Profesional Review. https://www.profesionalreview.com/2019/02/11/pcb-que-es/
